```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
from matplotlib_inline.backend_inline import set_matplotlib_formats
set_matplotlib_formats('retina')
```

## seaborn

```python
import seaborn as sns

tips = sns.load_dataset('tips')
```

## boxenplot

```python
sns.boxenplot(x=tips["total_bill"])
```

```python
sns.boxenplot(x="day", y="total_bill", data=tips)
```

```python
ax = sns.boxenplot(x="day", y="total_bill", hue="sex", data=tips,palette="coolwarm")
ax.legend(loc="upper left")
```

## pointplot

```python
sns.pointplot(x="day", y="total_bill", hue="sex",data=tips)

plt.ylim(0,30);
```

```python
sns.pointplot(x="day", y="total_bill", hue="sex",data=tips)
sns.barplot(x="day", y="total_bill", hue="sex",data=tips, alpha=0.5)
```

```python
tips["big_tip"] = (tips.tip / tips.total_bill) > 0.15
sns.lmplot(x="total_bill", y="big_tip", data=tips, y_jitter=.03);
```

```python
sns.lmplot(x="total_bill", y="big_tip", data=tips, logistic=True, y_jitter=.03)
```
